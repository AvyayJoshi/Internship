```python
#Case Study A: Employee Skill Tracker (HR Data Filtering)##
employees = {}

n = int(input("Enter number of employees: "))

for _ in range(n):
    name = input("Enter employee name: ")
    skills = input("Enter skills separated by commas: ").split(",")
    employees[name] = set(skill.strip() for skill in skills)

print("\nEmployees who know both Python and SQL:")
for name, skills in employees.items():
    if "Python" in skills and "SQL" in skills:
        print(name)

all_skills = set()
for skills in employees.values():
    all_skills.update(skills)

print("\nUnique skills across all employees:")
print(all_skills)

target = input("\nEnter employee name to compare skills: ")
target_skills = employees.get(target, set())

count = 0
for name, skills in employees.items():
    if name != target and target_skills & skills:
        count += 1

print("Number of employees sharing at least one skill with", target, ":", count)
```

```python
#Case Study B: Product Inventory Analyzer (Warehouse Comparison)
Warehouse1=[]
for _ in range(int(input("Enter number of products in Warehouse 1: "))):
    Warehouse1.append((
        input("Product ID: "),
        input("Product Name: "),
        int(input("Quantity: "))
    ))
Warehouse2=[]
for _ in range(int(input("Enter number of products in Warehouse 2: "))):
    Warehouse2.append((
        input("Product ID: "),
        input("Product Name: "),
        int(input("Quantity: "))
    ))

n1={i[1] for i in Warehouse1}
n2={i[1] for i in Warehouse2}

print("Common Products:")
print(*(n1 & n2),sep="\n")

print("Unique Products:")
print(*(n1 | n2),sep="\n")

print("All Products Sorted by Quantity:")
for i in sorted(Warehouse1+Warehouse2,key=lambda x:x[2],reverse=True):
    print(i)

```

```python
#Case Study C: University Course Registration System
students = {}
n = int(input("Enter number of students: "))
for _ in range(n):
    name = input("Student name: ")
    courses = tuple(input("Enter courses separated by commas: ").split(","))
    students[name] = tuple(c.strip() for c in courses)

print("\nStudents enrolled in both Data Science and AI:")
for name, courses in students.items():
    if "Data Science" in courses and "AI" in courses:
        print(name)

all_courses = set()
for courses in students.values():
    all_courses.update(courses)

print("\nTotal unique courses:")
print(len(all_courses))

unique_only = set()
for course in all_courses:
    count = sum(course in courses for courses in students.values())
    if count == 1:
        unique_only.add(course)

print("\nCourses not repeated by any student:")
print(unique_only)
```

```python
#Case Study D: Customer Purchase Insights (E-commerce Scenario)
records = []

n = int(input("Enter number of customers: "))
for _ in range(n):
    name = input("Customer name: ")
    products = input("Products purchased (comma separated): ").split(",")
    records.append((name, [p.strip() for p in products]))

all_products = set()
for r in records:
    all_products.update(r[1])

print("\nTotal unique products:")
print(len(all_products))

print("\nCustomers who purchased Laptop and Headphones:")
for name, products in records:
    if "Laptop" in products and "Headphones" in products:
        print(name)

freq = {}
for products in records:
    for p in products:
        freq[p] = freq.get(p, 0) + 1

max_freq = max(freq.values())
print("\nMost frequently purchased products:")
for k, v in freq.items():
    if v == max_freq:
        print(k)
```

```python
#Case Study E: Travel Planner (Data Deduplication Challenge)
bookings = []

n = int(input("Enter number of bookings: "))
for _ in range(n):
    name = input("Traveler name: ")
    dest = input("Destination: ")
    date = input("Date: ")
    bookings.append((name, dest, date))

unique_bookings = set(bookings)
print("\nUnique bookings:")
print(unique_bookings)

destinations = [b[1] for b in unique_bookings]
print("\nUnique destinations:")
print(set(destinations))

freq = {}
for d in destinations:
    freq[d] = freq.get(d, 0) + 1

top3 = sorted(freq.items(), key=lambda x: x[1], reverse=True)[:3]
print("\nTop 3 destinations:")
print(top3)
```

```python
#Case Study F: Student Attendance Analyzer
days = []

for i in range(5):
    students = set(input(f"Enter students present on Day {i+1} (comma separated): ").split(","))
    days.append(set(s.strip() for s in students))

present_all = set.intersection(*days)
print("\nStudents present every day:")
print(present_all)

all_students = set.union(*days)
absent_some = all_students - present_all

print("\nStudents absent on at least one day:")
print(absent_some)

print("\nTotal unique students:")
print(len(all_students))
```

```python
#Case Study G: Movie Recommendation Tracker
users = {}

n = int(input("Enter number of users: "))
for _ in range(n):
    name = input("User name: ")
    movies = tuple(input("Favorite movies (comma separated): ").split(","))
    users[name] = tuple(m.strip() for m in movies)

print("\nUsers sharing at least two movies:")
names = list(users.keys())
for i in range(len(names)):
    for j in range(i+1, len(names)):
        if len(set(users[names[i]]) & set(users[names[j]])) >= 2:
            print(names[i], names[j])

all_movies = set()
for m in users.values():
    all_movies.update(m)

print("\nAll unique movies:")
print(all_movies)

single_user_movies = set()
for movie in all_movies:
    count = sum(movie in m for m in users.values())
    if count == 1:
        single_user_movies.add(movie)

print("\nMovies liked by only one user:")
print(single_user_movies)
```

```python
# Case Study H: Sports Event Participation Summary
participants = {}

n = int(input("Enter number of participants: "))
for _ in range(n):
    name = input("Participant name: ")
    sports = input("Sports (comma separated): ").split(",")
    participants[name] = [s.strip() for s in sports]

print("\nParticipants in both Cricket and Football:")
for name, sports in participants.items():
    if "Cricket" in sports and "Football" in sports:
        print(name)

all_sports = set()
for s in participants.values():
    all_sports.update(s)

print("\nTotal unique sports:")
print(len(all_sports))

freq = {}
for s in participants.values():
    for sport in s:
        freq[sport] = freq.get(sport, 0) + 1

max_count = max(freq.values())
top_sports = tuple(s for s, c in freq.items() if c == max_count)

print("\nSports with highest participation:")
print(top_sports)
```
