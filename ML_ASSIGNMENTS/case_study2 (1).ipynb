```python
# Q3: Introduce 10% missing values randomly in Rating and Review columns
# Detect missing values and display the count of missing values per column
# Fill missing ratings with the median rating and missing reviews with "No review provided
import numpy as np

df = pd.DataFrame(data_dict)
num_missing = int(0.1 * len(df))
df.loc[np.random.choice(df.index, num_missing), 'Rating'] = np.nan
df.loc[np.random.choice(df.index, num_missing), 'Review'] = np.nan
print("Missing Values per Column:")
print(df.isnull().sum())
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df['Review'] = df['Review'].fillna("No review provided")
print("\nAfter Filling Missing Values:")
df.isnull().sum()
```

```python
# Q4: Create a new column 'TotalAmount' by multiplying Quantity and Price
# Add another column 'DiscountedPrice' that applies 10% discount on orders above 1000
# Use conditional selection to filter orders where TotalAmount > 15000

df['TotalAmount'] = df['Quantity'] * df['Price']

df['DiscountedPrice'] = df['TotalAmount'].where(
    df['TotalAmount'] <= 1000,
    df['TotalAmount'] * 0.9
)

filtered_df = df[df['TotalAmount'] > 15000]

print("Updated DataFrame:")
df


```

```python
# Q7: Sort the entire DataFrame by OrderDate (descending) and then by TotalAmount (descending)
# Create a new column 'SalesRank' that ranks products by TotalAmount in descending order

df_sorted = df.sort_values(
    by=['OrderDate', 'TotalAmount'],
    ascending=[False, False]
)

df_sorted['SalesRank'] = df_sorted['TotalAmount'].rank(
    method='dense',
    ascending=False
)

df_sorted
```

```python
# Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order
# Top 5 best-selling products based on total quantity sold
top_products = (
    df.groupby('Product')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_products
```

```python
# Q9: Group the data by 'Category' and calculate:
# - Total sales (sum of TotalAmount)
# - Average rating
# - Number of orders
# Display the results sorted by total sales
# Category-wise total sales, average rating, number of orders
category_stats = (
    df.groupby('Category')
    .agg(
        TotalSales=('TotalAmount', 'sum'),
        AvgRating=('Rating', 'mean'),
        Orders=('OrderID', 'count')
    )
    .sort_values(by='TotalSales', ascending=False)
)

category_stats
```

```python

# Q16: Create a correlation matrix between Quantity, Price, Rating, and TotalAmount using NumPy
# Calculate the covariance matrix
# Perform linear algebra operations: determinant and inverse of correlation matrix

# Q16: Correlation and covariance
data = df[['Quantity','Price','Rating','TotalAmount']].to_numpy()

corr = np.corrcoef(data, rowvar=False)
cov = np.cov(data, rowvar=False)

print(np.linalg.det(corr))
print(np.linalg.inv(corr))
```
