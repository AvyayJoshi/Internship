```python
# 3. Write a Python program to tokenize a given paragraph into sentences using sent_tokenize().

from nltk.tokenize import word_tokenize,sent_tokenize
text="It had been a simple realization that had changed Debra's life perspective. It was really so simple that she was embarrassed that she had lived the previous five years with the way she measured her worth. Now that she saw what she had been doing, she could see how sad it was. That made her all the more relieved she had made the change. The number of hearts her Instagram posts received wasn't any longer the indication of her own self-worth"
sentences=sent_tokenize(text)
print("sentence tokenization\n",sentences)

```

```python
# 4. Tokenize the same paragraph into individual words using word_tokenize().
from nltk.tokenize import word_tokenize
text="It had been a simple realization that had changed Debra's life perspective. It was really so simple that she was embarrassed that she had lived the previous five years with the way she measured her worth. Now that she saw what she had been doing, she could see how sad it was. That made her all the more relieved she had made the change. The number of hearts her Instagram posts received wasn't any longer the indication of her own self-worth"

word=word_tokenize(text)
print("word tokenization\n",word)
```

```python
# 5. Remove stopwords from a given text using NLTK stopwords list.

from nltk.corpus import stopwords

stop_words= set(stopwords.words('english'))
filter = [ w for w in word if w.lower() not in stop_words]
print("words after removing stop words .",filter)
```

```python
# 6. Perform stemming on a list of words using the PorterStemmer and print the output.

from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_words=[stemmer.stem(w) for w in filter]
print("stemmed words:\n",stemmed_words)
```

```python
# 7. Perform lemmatization on the same list of words using WordNetLemmatizer and compare results with stemming.

from nltk.stem import WordNetLemmatizer#add suffix to make meaning ful words

lemmatizer = WordNetLemmatizer()
lemmatizer_words = [lemmatizer.lemmatize(w) for w in filter]
print("lemmatization words:\n",lemmatizer_words)
```

```python
# 8. Apply Part of Speech (POS) tagging to a given sentence and print each word with its tag.

import nltk 
nltk.download('averaged_perceptron_tagger_eng')
pos_tags= nltk.pos_tag(filter)
print("Parts of speech Tagging:\n",pos_tags)
```

```python
# 9. Perform frequency distribution on a paragraph and print the top 10 most frequent words.
from nltk import FreqDist

freq_dist = FreqDist(filter)
print("Word Frequency:\n")

for word,freq in freq_dist.most_common(10):#show top 10
    print(f"{word}:{freq}")

freq_dist.plot(10,title="top 10 word frequencies")
import matplotlib.pyplot as plt
plt.savefig(' Assignmnet_plot.png',dpi=300,bbox_inches='tight')
```

```python
# 10. Plot a graph of the top 15 most frequent words using FreqDist().plot().
for word,freq in freq_dist.most_common(15):#show top 10
    print(f"{word}:{freq}")

freq_dist.plot(15,title="top 15 word frequencies")
import matplotlib.pyplot as plt
plt.savefig(' AssignmnetQ10_plot.png',dpi=300,bbox_inches='tight')
```

```python
# 16. Compare and print the difference between stemming and lemmatization for a list of 10 verbs.
verbs = ["running", "eating", "playing", "writing", "driving",
         "swimming", "reading", "talking", "jumping", "thinking"]

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Verb\t\tStemming\tLemmatization")

for verb in verbs:
    stem = stemmer.stem(verb)
    lemma = lemmatizer.lemmatize(verb, pos='v')
    print(f"{verb}\t\t{stem}\t\t{lemma}")
```
