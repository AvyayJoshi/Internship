```python
# Q3: Introduce 10% missing values randomly in Rating and Review columns
# Detect missing values and display the count of missing values per column
# Fill missing ratings with the median rating and missing reviews with "No review provided
import numpy as np

df = pd.DataFrame(data_dict)
num_missing = int(0.1 * len(df))
df.loc[np.random.choice(df.index, num_missing), 'Rating'] = np.nan
df.loc[np.random.choice(df.index, num_missing), 'Review'] = np.nan
print("Missing Values per Column:")
print(df.isnull().sum())
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df['Review'] = df['Review'].fillna("No review provided")
print("\nAfter Filling Missing Values:")
df.isnull().sum()
```

```python

# Q27: Create three separate DataFrames for three different months of sales data
# Concatenate them vertically (along rows) and horizontally (along columns)
# Remove duplicate records if any and reset the index
# Q27: Monthly DataFrames, concatenation, duplicates handling
import pandas as pd

df['OrderDate'] = pd.to_datetime(df['OrderDate'])

jan = df[df['OrderDate'].dt.month == 1]
feb = df[df['OrderDate'].dt.month == 2]
mar = df[df['OrderDate'].dt.month == 3]

vertical_concat = pd.concat([jan, feb, mar])

horizontal_concat = pd.concat([jan.reset_index(drop=True),
                               feb.reset_index(drop=True),
                               mar.reset_index(drop=True)], axis=1)

final_df = vertical_concat.drop_duplicates().reset_index(drop=True)

print(final_df)
```

```python

# Q28: Convert OrderDate column to datetime format
# Extract: Year, Month, Day, DayOfWeek, Quarter from OrderDate
# Create a time series analysis showing sales trends by month and quarter
# Calculate rolling 7-day average of sales
# Q28: Date conversion, feature extraction, time series analysis
import pandas as pd
import matplotlib.pyplot as plt

df['OrderDate'] = pd.to_datetime(df['OrderDate'])

df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['Day'] = df['OrderDate'].dt.day
df['DayOfWeek'] = df['OrderDate'].dt.day_name()
df['Quarter'] = df['OrderDate'].dt.quarter

monthly_sales = df.groupby(df['OrderDate'].dt.to_period('M'))['TotalAmount'].sum()
quarterly_sales = df.groupby(df['OrderDate'].dt.to_period('Q'))['TotalAmount'].sum()

daily_sales = df.sort_values('OrderDate').set_index('OrderDate')['TotalAmount']
rolling_avg = daily_sales.rolling(window=7).mean()

monthly_sales.plot(title='Monthly Sales Trend')
plt.show()

quarterly_sales.plot(title='Quarterly Sales Trend')
plt.show()

rolling_avg.plot(title='7-Day Rolling Average of Sales')
plt.show()
```

```python

# Q29: Using the Review column, perform comprehensive text analysis:
# - Tokenize all reviews into words
# - Remove stopwords from reviews
# - Apply stemming and lemmatization
# - Perform POS tagging on sample reviews
# - Create frequency distribution of top 20 most common words
# - Visualize word frequency using a bar chart
# Q29: Comprehensive text analysis on Review column
import nltk
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, FreqDist

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

text = " ".join(df['Review']).lower()

tokens = word_tokenize(text)

filtered = [w for w in tokens if w.isalpha() and w not in stopwords.words('english')]

stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered]

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]

pos_tags = pos_tag(tokens[:10])
print(pos_tags)

freq = FreqDist(lemmatized)
top_words = freq.most_common(20)
print(top_words)

words, counts = zip(*top_words)
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.show()
```
