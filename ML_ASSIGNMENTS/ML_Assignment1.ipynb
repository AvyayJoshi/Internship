```python
df = pd.DataFrame({
    "Student_Name": ["Aman", "Riya", "Karan", "Sneha", "Vijay"],
    "Maths": [70, 85, 90, 65, 88],
    "Science": [72, 88, 91, 68, 90],
    "English": [75, 80, 78, 70, 85]
})

print(df.head(3))
df
```

```python
df.to_csv("students.csv", index=False)

df2 = pd.read_csv("students.csv")
print(df2.shape)
print(df2.columns)
```

```python
df["Total_Marks"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total_Marks"] / 3
print(df)
```

```python
import numpy as np

df.loc[1, "Science"] = np.nan

print(df.isnull())

df["Science"] = df["Science"].fillna(df["Science"].mean())
print(df)
```

```python
print(df[df["Average"] > 75])
```

```python
print(df.sort_values("Total_Marks", ascending=False).head(3))
```

```python
df["English_Group"] = ["Above 70", "Above 70", "Above 70", "Below 70", "Above 70"]

print(df.groupby("English_Group")[["Maths", "Science"]].mean())
```

```python
df_marks = df[["Student_Name", "Total_Marks"]]

df_sports = pd.DataFrame({
    "Student_Name": ["Aman", "Riya", "Karan", "Sneha", "Vijay"],
    "Sports_Score": [20, 18, 22, 15, 25]
})

final_df = pd.merge(df_marks, df_sports, on="Student_Name")
final_df["Overall_Score"] = final_df["Total_Marks"] + final_df["Sports_Score"]

print(final_df)
```

```python
final_df.to_csv("final_student_scores.csv", index=False)
print(" DATA SAVED SUCCESSFULY")
```
