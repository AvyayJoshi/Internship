```python
# 1. Students registered for Coding Hackathon (Set A)
hackathon_students = {"Amit", "Priya", "Rahul", "Sneha", "Vikram"}

# 2. Students registered for Robotics Workshop (Set B)
robotics_students = {"Rahul", "Karan", "Sneha", "Pooja", "Arjun"}

# 3. Gate Entry Logs (List with duplicates)
# Students swiped their ID cards to enter (some swiped twice by mistake)
entry_logs = ["ID_101", "ID_102", "ID_101", "ID_103", "ID_102"]

# 4. VIP Guests (Frozen Set - Immutable)
vip_guests = frozenset(["Dean", "Principal", "Chief Guest"])

print("Base Dataset ")
print("Hackathon Team:", hackathon_students)
print("Robotics Team:", robotics_students)
```

```python
# Problem 1: Find the list of all unique students participating in at least one event.
# Concept: Union (|) combines both sets. 

all_participants = hackathon_students | robotics_students
# OR: hackathon_students.union(robotics_students)

print("Total Unique Participants:", all_participants)
```

```python
# Problem 2: Find students who are participating in BOTH events.
# Concept: Intersection (&) finds common elements. 

common_students = hackathon_students & robotics_students
# OR: hackathon_students.intersection(robotics_students)

print("Students in both events:", common_students)
```

```python
# Problem 3: Find students who are ONLY in the Hackathon (not in Robotics).
# Concept: Difference (-) removes elements of the second set from the first. 

only_hackathon = hackathon_students - robotics_students

print("Students only in Hackathon:", only_hackathon)
```

```python
# Problem 4: Find students who are in either Hackathon or Robotics, but NOT both.
# Concept: Symmetric Difference (^) removes the common intersection. 

single_interest = hackathon_students ^ robotics_students

print("Students with only one interest:", single_interest)
```

```python
# Problem 5: The 'entry_logs' list has duplicates because some students swiped twice. 
# We need a list of unique attendees.
# Concept: Converting a List to a Set automatically removes duplicates.

unique_attendees = set(entry_logs)

print(f"Raw Entry Count: {len(entry_logs)}")
print(f"Actual Unique Attendees: {len(unique_attendees)}")
print("Unique IDs:", unique_attendees)
```

```python
# Problem 6: A new student 'Zoya' wants to join the Hackathon.
# Concept: .add() adds a single element. 

# Creating a copy to keep original dataset clean for other cells
hackathon_updated = hackathon_students.copy()
hackathon_updated.add("Zoya")

print("Updated Hackathon List:", hackathon_updated)
```

```python
# Problem 7: 'Amit' decided to leave the Hackathon. Remove him.
# Concept: .remove() deletes a specific element. 

hackathon_updated.remove("Amit")

print("List after Amit left:", hackathon_updated)
```

```python
# Problem 8: Check if 'Rahul' is registered for the Robotics Workshop.
# Concept: 'in' keyword is used for membership testing. 

is_registered = "Rahul" in robotics_students

print("Is Rahul in Robotics?", is_registered)
```

```python
# Problem 9: How many students are in the Robotics team?
# Concept: len() counts the elements. 

team_size = len(robotics_students)

print(f"Robotics Team Size: {team_size}")
```

```python
# Problem 10: Try to add a name to the VIP Guest list (Frozen Set).
# Concept: Frozen sets are immutable and will raise an error if modified. 

try:
    vip_guests.add("Hacker")
except AttributeError as e:
    print("Error: Cannot modify VIP list!", e)
```
