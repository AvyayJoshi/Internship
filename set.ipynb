```python
# Creating a set with integer elements
numbers = {1, 2, 3, 4, 5}
numbers
```

```python
# Creating a set using set() constructor
letters = set(["a", "b", "c"])
letters
```

```python
# Creating an empty set ({} creates a dictionary)
empty_set = set()
empty_set
```

```python
# Duplicate elements are automatically removed in a set
dup = {1, 2, 2, 3, 3, 4}
dup
```

```python
# Checking whether an element exists in the set
3 in numbers
```

```python
# Finding number of elements in the set
len(numbers)
```

```python
# Adding a single element to the set
numbers.add(6)
numbers
```

```python
# Adding multiple elements to the set
numbers.update([7, 8, 9])
numbers
```

```python
# Removing a specific element (throws error if not found)
numbers.remove(9)
numbers
```

```python
# Removing an element without error if element does not exist
numbers.discard(10)
numbers
```

```python
# Removing and returning a random element from the set
numbers.pop()
numbers
```

```python
# Removing all elements from the set
temp = {1, 2, 3}
temp.clear()
temp
```

```python
# Union of two sets (all unique elements)
a = {1, 2, 3}
b = {3, 4, 5}
a.union(b)
```

```python
# Intersection of two sets (common elements)
a.intersection(b)
```

```python
# Elements present in set a but not in set b
a.difference(b)
```

```python
# Elements present in either set but not in both
a.symmetric_difference(b)
```

```python
# Union using operator
a | b
```

```python
# Intersection using operator
a & b
```

```python
# Difference using operator
a - b
```

```python
# Symmetric difference using operator
a ^ b
```

```python
# Checking whether all elements of c are in a
c = {1, 2}
c.issubset(a)
```

```python
# Checking whether set a contains all elements of c
a.issuperset(c)
```

```python
# Checking whether two sets have no common elements
x = {10, 11}
a.isdisjoint(x)
```

```python
# Creating a shallow copy of a set
copy_set = a.copy()
copy_set
```

```python
# Creating an immutable set (cannot be modified)
fs = frozenset([1, 2, 3])
fs
```

```python
#Returns True if at least one element in the set evaluates to True.
s = {0, 0, 1}
any(s)
```

```python
#Returns True if all elements in the set evaluate to True.
all(s)
```

```python
#to find the minimum element  of the set
marks = {78, 85, 92}
min(marks)
```

```python
# to find the maximum of the element
max(marks)
```

```python
# Creating a set of squares using set comprehension
squares = {x * x for x in range(1, 6)}
squares
```

```python
# Creating set of squares of even numbers only
even_squares = {x * x for x in range(1, 11) if x % 2 == 0}
even_squares
```

```python
# Students enrolled in Python and Java courses
python_students = {"Aman", "Riya", "Kunal", "Neha"}
java_students = {"Riya", "Neha", "Arjun", "Sonal"}
```

```python
# Students enrolled in both Python and Java
python_students & java_students
```

```python
# Students enrolled in at least one course
python_students | java_students
```

```python
# Students enrolled only in Python
python_students - java_students
```

```python
# Students enrolled only in Java
java_students - python_students
```

```python
# Students enrolled in exactly one course
python_students ^ java_students
```

```python
# Checking if all Python students are also Java students
python_students.issubset(java_students)
```

```python
# Adding a new student to Python course
python_students.add("Mehul")
python_students
```

```python
# Removing a student who dropped the Java course
java_students.remove("Arjun")
java_students
```

```python
# Final enrollment summary using sets
{
    "Python": python_students,
    "Java": java_students,
    "Both": python_students & java_students,
    "Total Unique Students": python_students | java_students
}
```
