```python
# Q1. Print numbers from 1 to 10
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
# Q5. Reverse a given number
num = int(input("Enter number for reverse: "))
rev = 0
temp = num
while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10
print("Reversed number:", rev)
```

```python
num = int(input("Enter number: "))
count = 0
while num > 0:
    count += 1
    num //= 10
print("Digits:", count)
```

```python
# Q7. Factorial using while loop
num = int(input("Enter number for factorial: "))
fact = 1
while num > 0:
    fact *= num
    num -= 1
print("Factorial:", fact)
```

```python
# Q8. Fibonacci sequence up to 50
a, b = 0, 1
while a <= 50:
    print(a, end=" ")
    a, b = b, a + b
```

```python
# Q9. Palindrome check
num = int(input("Enter number to check palindrome: "))
temp = num
rev = 0
while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10
if num == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Q6. Sum of digits
num = int(input("Enter number to sum digits: "))
sum_digits = 0
while True:
    sum_digits += num % 10
    num //= 10
    if num == 0:
        break
print("Sum of digits:", sum_digits)
```

```python
# Q8. Prime number check
num = int(input("Enter number to check prime: "))
i = 2
is_prime = True
while True:
    if i >= num:
        break
    if num % i == 0:
        is_prime = False
        break
    i += 1
print("Prime" if is_prime and num > 1 else "Not Prime")
```
